# Artificial Analysis Intelligence Index (AAII) of different AI Models


### A. Artificial Analysis Intelligence Index 
- **Artificial Analysis Intelligence Index** is a measure of evaluating AI models considering problem-solving skills, creativity, accuracy, reasoning abilities, understanding of contexts, etc. It is helpful to choose and compare Large Language models (LLMs) like Gemini, Claude, Gpt-4, etc. If a model performs high on each this, it is generally considered as the better generator for relevant facts and figures like correct answers, descriptions, better problem solving and analytical abilities, etc. It basically evaluates the general knowledge, reasoning, mathematical problem-solving, and coding capabilities on these models.

### B. Artificial Analysis Coding Index
-
  **Artificial Analysis Coding Index** is a measure of evaluating programmatical abilities of an AI model. It evaluates coding, debugging and understanding patterns of a system's program. It is crucial to know the performance effectiveness of different AI models that can help us on assisting for software development, automation and script generation tasks. Models like GPT-4, Code Llama, or Codex are often rated high on coding indices because they handle multiple languages, algorithms, and debugging tasks well.

### C. Artificial Analysis Math Index
- **Artificial Analysis Math Index** is a measure of evaluating an AI model's tendency of solving mathematical problems, perform calculations without any error and provide quantitative reasoning on mathematical problems. An Artificial Analysis Math Index is critical for applications like data science, quantitative research, financial modeling, or AI reasoning. These models are rated on their ability to solve algebra, calculus, probability, and logic problems without error.

```python
import pandas as pd
from tabulate import tabulate
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## A. Viewing the dataset
- Initially each relevant column and row values in the following dataset about the leadership of AI models with context to given indexes is read below.

```python
df=pd.read_csv("modelData.csv",low_memory=False)
df
```

### I. Visualization of Artificial Analysis Intelligence Index with a Bar Graph
- The bar chart below shows the Artificial Analysis Intelligence Index for each model, sorted from highest to lowest.

```python
# Bar Chart for Artificial Analysis Intelligence Index
plt.figure(figsize=(12, 8))
df_sorted = df.sort_values(by='Artificial Analysis Intelligence Index', ascending=False)
sns.barplot(
    x='Artificial Analysis Intelligence Index',
    y='Model',
    data=df_sorted,
    hue='Model'
)
plt.title('Artificial Analysis Intelligence Index of AI Models', fontsize=16)
plt.xlabel('Intelligence Index', fontsize=12)
plt.ylabel('AI Model', fontsize=12)
plt.tight_layout()
plt.savefig('intelligence_index_seaborn.png')
plt.show()
```

### II. Visualization of Artificial Analysis Coding Index with a Bar Graph
- The bar chart below shows the Artificial Analysis Coding Index for each model, sorted from highest to lowest.

```python
# Bar Chart for Artificial Analysis Coding Index
plt.figure(figsize=(12, 8))
df_sorted = df.sort_values(by='Coding Index', ascending=False)
sns.barplot(
    x='Coding Index',
    y='Model',
    data=df_sorted,
    hue='Model'
)
plt.title('Artificial Analysis Coding Index of AI Models', fontsize=16)
plt.xlabel('Coding Index', fontsize=12)
plt.ylabel('AI Model', fontsize=12)
plt.tight_layout()
plt.savefig('coding_index_seaborn.png')
plt.show()
```

### III. Visualization of Artificial Analysis Math Index with a Bar Graph
- The bar chart below shows the Artificial Analysis Math Index for each model, sorted from highest to lowest.

```python
# Bar Chart for Artificial Analysis Coding Index
plt.figure(figsize=(12, 8))
df_sorted = df.sort_values(by='Math Index', ascending=False)
sns.barplot(
    x='Math Index',
    y='Model',
    data=df_sorted,
    hue='Model'
)
plt.title('Artificial Analysis Math Index of AI Models', fontsize=16)
plt.xlabel('Math Index', fontsize=12)
plt.ylabel('AI Model', fontsize=12)
plt.tight_layout()
plt.savefig('math_index_seaborn.png')
plt.show()
```

### IV. Visualization for Intelligence Index vs. Price of these AI models
- This analysis helps us determine the value of an AI model. It's a comparison between a model's Intelligence Index (a measure of its overall performance) and its cost (usually measured in USD per million tokens).

```python
# 2. Scatter Plot for Intelligence vs. Price
# Calculate the average token cost
df['Average Token Cost'] = (df['Input USD per 1M Tokens'] + df['Output USD per 1M Tokens']) / 2
plt.figure(figsize=(14, 8))
sns.scatterplot(
    x='Average Token Cost',
    y='Artificial Analysis Intelligence Index',
    data=df,
    hue='Model',
    size='Context Window',
    sizes=(50, 500),
    style='Creator'
)
plt.title('Intelligence vs. Price (Average Token Cost)', fontsize=16)
plt.xlabel('Average Token Cost (USD per 1M Tokens)', fontsize=12)
plt.ylabel('Artificial Analysis Intelligence Index', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('intelligence_vs_price.png')
plt.show()
```

### V.  Intelligence vs. Output Speed
- This analysis compares a model's Intelligence Index with its Median Tokens per Second (a measure of how fast it generates a response). This helps us understand the trade-off between quality and speed.

```python
# 5. Scatter Plot for Intelligence vs. Output Speed
df_cleaned = df.dropna(subset=['Median Tokens per s'])
plt.figure(figsize=(14, 8))
sns.scatterplot(
    x='Median Tokens per s',
    y='Artificial Analysis Intelligence Index',
    data=df_cleaned,
    hue='Model',
    size='Context Window',
    sizes=(50, 500),
    style='Creator'
)
plt.title('Intelligence vs. Output Speed of AI Models', fontsize=16)
plt.xlabel('Median Tokens per Second', fontsize=12)
plt.ylabel('Artificial Analysis Intelligence Index', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('intelligence_vs_output_speed.png')
plt.show()
```

### VI. Context window 
- The context window size is a crucial technical specification for AI models that refers to the amount of text, or tokens, an AI can process and remember at once.
It's the total number of tokens (words, parts of words, or characters) that the model can see and consider when generating its next response. A token is a fundamental unit of language that models use to process information.

```python
# 3. Bar Chart for Context Window Size
df_context = df.sort_values(by='Context Window', ascending=False)
plt.figure(figsize=(14, 8))
sns.barplot(
    x='Context Window',
    y='Model',
    data=df_context,
)
plt.title('Context Window Size of AI Models', fontsize=16)
plt.xlabel('Context Window Size (Tokens)', fontsize=12)
plt.ylabel('AI Model', fontsize=12)
plt.tight_layout()
plt.savefig('context_window_size.png')
plt.show()
```
